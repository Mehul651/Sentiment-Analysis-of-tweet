```python
#Data cleaning 
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download the 'punkt' resource
nltk.download('punkt')

# Assuming you have already loaded your dataset into a DataFrame called df

# Function to clean a single text
def clean_text(text):
    # Remove special characters, URLs, and user mentions
    text = re.sub(r"http\S+|www\S+|@[^\s]+", "", text)
    
    # Remove punctuation and numbers
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Tokenize the text
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Join the tokens back into a clean text
    clean_text = " ".join(tokens)
    
    return clean_text

# Apply the cleaning function to the 'text' column
df['clean_text'] = df['text'].apply(clean_text)

# Display the cleaned DataFrame
print(df[['text', 'clean_text']])
```

```python
#Top 20 most common words along with their frequencies

from collections import Counter

# Concatenate all the cleaned text into a single string
all_cleaned_text = ' '.join(df['clean_text'])

# Tokenize the text into individual words
words = all_cleaned_text.split()

# Count the frequency of each word
word_counts = Counter(words)

# Sort the words based on their frequencies
sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# Display the top 20 most common words
top_20_words = sorted_word_counts[:20]
print(top_20_words)
```

```python
#Bernoulli model
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Splitting the data
X = df['clean_text']  # Features (cleaned text)
y = df['sentiment']   # Target (sentiment)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=0)

# Initialize the TF-IDF vectorizer with unigrams and bigrams, limiting features to 500,000
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=500000)

# Fit the vectorizer to the training data and transform the training and testing data
X_train_transformed = tfidf_vectorizer.fit_transform(X_train)
X_test_transformed = tfidf_vectorizer.transform(X_test)

# Create and fit the Bernoulli Naive Bayes model
bnb_model = BernoulliNB()
bnb_model.fit(X_train_transformed, y_train)

# Evaluate the model
y_pred_bnb = bnb_model.predict(X_test_transformed)
print(classification_report(y_test, y_pred_bnb))

# Compute and plot the confusion matrix
cf_matrix_bnb = confusion_matrix(y_test, y_pred_bnb)

categories = ['Negative', 'Positive']
group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix_bnb.flatten() / np.sum(cf_matrix_bnb)]

labels_bnb = [f'{v1}\n{v2}' for v1, v2 in zip(group_names, group_percentages)]
labels_bnb = np.asarray(labels_bnb).reshape(2, 2) 


# Plot the confusion matrix
sns.heatmap(cf_matrix_bnb, annot=labels_bnb, cmap='GnBu', fmt='',
            xticklabels=categories, yticklabels=categories)

plt.xlabel("Predicted values", fontdict={'size': 14}, labelpad=10)
plt.ylabel("Actual values", fontdict={'size': 14}, labelpad=10)
plt.title("Confusion Matrix (Bernoulli Naive Bayes)", fontdict={'size': 18}, pad=20)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import joblib
from tkinter import Tk, simpledialog

df = pd.read_csv("C:/Users/mehul/clean_text.csv")

# Drop rows with missing values
df.dropna(subset=['clean_text'], inplace=True) 

# Splitting the data
X = df['clean_text']  # Features (cleaned text)
y = df['sentiment']   # Target (sentiment)

# Adding Neutral sentiment
y[y == 'positive'] = 1
y[y == 'negative'] = 0
y[y == 'neutral'] = 2

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=0)

# Initialize the TF-IDF vectorizer with unigrams and bigrams, limiting features to 500,000
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), max_features=500000)

# Fit the vectorizer to the training data and transform the training and testing data
X_train_transformed = tfidf_vectorizer.fit_transform(X_train)
X_test_transformed = tfidf_vectorizer.transform(X_test)

# Create and fit the Bernoulli Naive Bayes model
bnb_model = BernoulliNB()
bnb_model.fit(X_train_transformed, y_train)

# Save Bernoulli Naive Bayes model
joblib.dump(bnb_model, 'bernoulli_naive_bayes_model.pkl')

# Create and fit the LinearSVC model
svc_model = LinearSVC()
svc_model.fit(X_train_transformed, y_train)

# Save LinearSVC model
joblib.dump(svc_model, 'linear_svc_model.pkl')

# Create and fit the Logistic Regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_train_transformed, y_train)

# Save Logistic Regression model
joblib.dump(logistic_model, 'logistic_regression_model.pkl')

# Function to predict sentiment
def predict_sentiment(sentence):
    models = {
        'Bernoulli Naive Bayes': bnb_model,
        'LinearSVC': svc_model,
        'Logistic Regression': logistic_model
    }
    results = {}
    for name, model in models.items():
        prediction = model.predict(tfidf_vectorizer.transform([sentence]))[0]
        if prediction == 1:
            results[name] = 'Positive'
        elif prediction == 0:
            results[name] = 'Negative'
        else:
            results[name] = 'Neutral'
    return results

# GUI for input sentence and displaying sentiment prediction
def main():
    root = Tk()
    root.withdraw()  # Hide the root window

    while True:
        sentence = simpledialog.askstring("Input", "Enter a sentence (or type 'exit' to quit):")
        if sentence.lower() == 'exit':
            break
        else:
            results = predict_sentiment(sentence)
            print("Sentiment Analysis Results:")
            for model, sentiment in results.items():
                print(f"{model}: {sentiment}")

if __name__ == "__main__":
    main()
```
